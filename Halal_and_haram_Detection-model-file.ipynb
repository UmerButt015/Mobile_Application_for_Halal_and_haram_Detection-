```python
# Get class names from data.yaml to create labels.txt
import yaml

# Read the data.yaml file
data_yaml_path = '/content/halal-logo-1/data.yaml'
if os.path.exists(data_yaml_path):
    with open(data_yaml_path, 'r') as f:
        data = yaml.safe_load(f)

    # Extract class names
    if 'names' in data:
        class_names = data['names']
        print("ğŸ“‹ Class names found:")
        for i, name in enumerate(class_names):
            print(f"  {i}: {name}")

        # Create labels.txt content
        labels_content = '\n'.join(class_names)
        print("\nğŸ“„ Labels.txt content:")
        print(labels_content)

        # Save and download labels.txt
        with open('/content/labels.txt', 'w') as f:
            f.write(labels_content)

        print("\nğŸ“¥ Downloading labels.txt...")
        files.download('/content/labels.txt')
    else:
        print("âŒ No class names found in data.yaml")
else:
    print("âŒ data.yaml file not found")
```
